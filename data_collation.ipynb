```python
import pandas as pd
import urllib2
from bs4 import BeautifulSoup
```

```python
website = "https://www.moneycontrol.com"
url_sector = "/india/stockmarket/sector-classification/marketstatistics/bse/agri.html"
```

```python
context = ssl._create_unverified_context()
response = urllib2.urlopen(website+url_sector, context=context)
html_doc = response.read()
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
# soup = soup.get_text()
#print(soup.find("table"),{"class":"tbldata14 bdrtpg"})
# for table in soup.find_all('table'):
#     print(table.get('class'))
data_table = soup.find('table', class_="tbldata14")
# print(data_table)
```

```python
#df = pd.read_html(website+url_sector)
# print(data_table.find_all('tr'))
```

```python
df = pd.DataFrame(columns=['Company Name','Industry','Last Price','Change','%Chg','Mkt Cap'])
```

```python
Company_name = []
Industry = []
Last_Price = []
Change = []
Change_prcnt = []
Market_Cap = []
Company_link = []
for row in data_table.find_all('tr'):
    data_columns = row.find_all('td')
    
    if data_columns != []:
        # print(data_columns[0].text.strip())
        Company_name.append(data_columns[0].text.strip())
        Link = data_columns[0].find('a').get('href')
        Company_link.append(website + Link)
        Industry.append(data_columns[1].text.strip())
        Last_Price.append(data_columns[2].text.strip())
        Change.append(data_columns[3].text.strip())
        Change_prcnt.append(data_columns[4].text.strip())
        Market_Cap.append(data_columns[5].text.strip())  
        # df = df.append({'Company Name':data_columns[0].text.strip(),'Industry':Industry,'Last Price':Last_Price,'Change':Change,'%Chg':Change_prcnt,'Mkt Cap':Market_Cap}, ignore_index=True)
```
